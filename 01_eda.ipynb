```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading csv to Dataframe
df = pd.read_csv('manufacturing_defect_dataset.csv')

# Show first rows of Dataframe
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
selected_columns = [
    'ProductionVolume', 'SupplierQuality', 'MaintenanceHours', 
    'EnergyConsumption', 'DowntimePercentage', 'WorkerProductivity', 
    'AdditiveProcessTime', 'DefectRate'
]

df[selected_columns].describe().round(2)
```

```python
selected_columns = [
    'DefectRate'
]

df[selected_columns].describe().round(2)
```

```python
df['MaintenanceHoursPerDay'] = df['MaintenanceHours'] / 7

selected_columns = [
    'ProductionVolume', 'SupplierQuality', 'MaintenanceHoursPerDay'
]

Descriptive_statisticsTable = df[selected_columns].describe().round(2)
Descriptive_statisticsTable
```

```python
# Export the descriptive statistics table to LaTeX format
latex_table_code = Descriptive_statisticsTable.to_latex(index=True, float_format="%.2f")
print(latex_table_code)
```

# Correlation Matrix

```python
# Calculate correlation matrix
correlation_matrix = df[['ProductionVolume', 'SupplierQuality', 'MaintenanceHoursPerDay', 
                              'EnergyConsumption', 'DowntimePercentage', 'WorkerProductivity', 
                              'AdditiveProcessTime', 'DefectRate', 'DefectStatus']].corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Key Variables')
plt.savefig('Figures/Figures_EDA/EDA_correlationmatrix.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Distribution Analysis Target Variable (DefectRate)

```python
# Plot the distribution of DefectRate
plt.figure(figsize=(8, 6))
sns.histplot(df['DefectRate'], kde=True)
plt.title('Distribution of DefectRate')
plt.xlabel('Defect Rate')
plt.ylabel('Frequency')
plt.savefig('Figures/Figures_EDA/EDA_TargetDistribution.png', dpi=300)
plt.show()
```

# Scatter Plots with Defect Rate (Bivariate Analysis)

```python
# Scatter plot for Production Volume vs Defect Rate
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['ProductionVolume'], y=df['DefectRate'])
plt.title('Production Volume vs Defect Rate')
plt.xlabel('Production Volume')
plt.ylabel('Defect Rate')
plt.show()

# Scatter plot for Supplier Quality vs Defect Rate
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['SupplierQuality'], y=df['DefectRate'])
plt.title('Supplier Quality vs Defect Rate')
plt.xlabel('Supplier Quality')
plt.ylabel('Defect Rate')
plt.show()

# Scatter plot for Maintenance Hours (Per Day) vs Defect Rate
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['MaintenanceHoursPerDay'], y=df['DefectRate'])
plt.title('Maintenance Hours (Per Day) vs Defect Rate')
plt.xlabel('Maintenance Hours (Per Day)')
plt.ylabel('Defect Rate')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['ProductionVolume'], y=df['EnergyConsumption'], color='blue')

plt.title('Ratio between ProductionVolume and EnergyConsumption')
plt.xlabel('ProductionVolume')
plt.ylabel('EnergyConsumption')
plt.show()
```

```python
# Box plot for Production Volume
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['ProductionVolume'])
plt.title('Box Plot of Production Volume')
plt.ylabel('Production Volume')
plt.show()

# Box plot for Supplier Quality
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['SupplierQuality'])
plt.title('Box Plot of Supplier Quality')
plt.ylabel('Supplier Quality')
plt.show()

# Box plot for Maintenance Hours (Per Day)
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['MaintenanceHoursPerDay'])
plt.title('Box Plot of Maintenance Hours (Per Day)')
plt.ylabel('Maintenance Hours (Per Day)')
plt.show()
```

# EDA Findings
Correlation Analysis: The correlation analysis revealed no significant correlations between the variables and defect rate. This lack of correlation suggests that defect rate may be influenced by more complex, non-linear relationships that are not captured by simple linear correlations.

Distribution of Defect Rate: The defect rate showed an even distribution across samples, with no apparent normal distribution or skewness. This balanced distribution implies that defect rate values are spread uniformly, which could be beneficial for model stability, as the target variable does not exhibit extreme biases.

Scatter Plots: No visible trends or relationships were identified between defect rate and key variables, including production volume, supplier quality, and maintenance hours. This lack of trend suggests that straightforward linear or non-linear trends may not fully capture the relationship between defect rates and other variables.

Outlier Detection: No outliers were detected in the dataset, indicating consistent data entries across observations. The absence of outliers simplifies the data preparation process and suggests that extreme production conditions or anomalies are not present in this dataset. This result was anticipated, as the dataset is synthetic, designed to represent typical production scenarios without irregularities.
