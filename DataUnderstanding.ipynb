# Testing for imbalance data

```python
# Summary statistics for DefectRate
skewness = df['DefectRate'].skew()
kurtosis = df['DefectRate'].kurtosis()

print(f'Skewness: {skewness}')
print(f'Kurtosis: {kurtosis}')
df['DefectRate'].quantile([0.1, 0.25, 0.5, 0.75, 0.9, 0.95, 0.99])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of DefectRate
plt.figure(figsize=(10, 6))
sns.histplot(df['DefectRate'], bins=30, kde=True)
plt.title('Distribution of DefectRate')
plt.xlabel('Defect Rate')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Count occurrences in defined bins
bins = [0, 1, 2, 3, 4, 5]
labels = ['0-1', '1-2', '2-3', '3-4', '4-5']
df['DefectRateBin'] = pd.cut(df['DefectRate'], bins=bins, labels=labels, right=False)
bin_counts = df['DefectRateBin'].value_counts().sort_index()

# Display counts in each bin
print("Counts in Each DefectRate Bin:")
print(bin_counts)
```

# Checking for missing values

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Check for missing values in the expanded features
features = ['DefectRate', 'ProductionVolume', 'SupplierQuality', 'MaintenanceHours', 'MaintenanceHoursPerDay', 'EnergyConsumption']
missing_values = df[features].isnull().sum()
print("Missing Values in Each Column:")
print(missing_values)

# Set the plot size
plt.figure(figsize=(12, 10))

# Create boxplots for the expanded set of features
for i, feature in enumerate(features):
    plt.subplot(3, 2, i + 1)  # Adjust subplot for better layout
    sns.boxplot(data=df[feature])
    plt.title(f'Boxplot of {feature}')

plt.tight_layout()
plt.show()
```
